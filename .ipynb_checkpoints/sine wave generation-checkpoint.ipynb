## Current config for Oxford

```python
#configure X galvo
rate = 400000
freq = 100
Xamplitude = 0.20
Xoffset = 1.175
Yamplitude = 0.015
Yoffset = 0.05
Xdelay = .01
Ydelay = .5
time = np.linspace(0, 2*math.pi, round(rate/freq))
Xsamples = Xamplitude*np.sin(time) + Xoffset
Ysamples = Yamplitude*np.sin(time) + Yoffset
position = np.vstack((Xsamples,Ysamples)) #appends second row to samples for Y galvo

Xtime = time + Xdelay
Ytime = time + Ydelay
Xsamples = Xamplitude*np.sin(Xtime) + Xoffset
Ysamples = Yamplitude*np.sin(Ytime) + Yoffset
position_with_delay = np.vstack((Xsamples,Ysamples)) #appends second row to samples for Y galvo
```

```python
plt.figure()
for time in range(position.shape[1]):
    x = position[0,time]
    y = position[1, time]
    plt.scatter(x,y)
    x = position_with_delay[0,time]
    y = position_with_delay[1, time]
    plt.scatter(x,y)
    
plt.figure()
plt.plot(position[0,:])
plt.plot(position_with_delay[0,:])

plt.figure()
plt.plot(position[1,:])
plt.plot(position_with_delay[1,:])
```
